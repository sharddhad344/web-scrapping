# Find all movies data

```python
names,ratings,metas,votes,gross=[],[],[],[],[]

for i in range(1,1001,50):
    url=f"https://www.imdb.com/search/title/?release_date=2017-01-01,2017-12-31&sort=num_votes,desc&start={i}&ref_=adv_nxt"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html5lib')
    all_movies=soup.find_all(class_="lister-item-content")
    for f in all_movies:
        if f.find(class_="metascore"): 
            names.append(f.find(class_="lister-item-header").get_text().replace('\n','').replace(' ',''))
            ratings.append(f.find(class_="inline-block ratings-imdb-rating").get_text().strip())
            metas.append(f.find(class_="metascore").get_text().strip())

            fvg=f.find(class_="sort-num_votes-visible").get_text().replace('\n','').replace(' ','').split("|")
            if len(fvg)==2:
                votes.append(fvg[0].split(':')[1].replace(',', ''))
                gross.append(fvg[1].split(':')[1])
            else:
                votes.append(fvg[0].split(':')[1].replace(',', ''))
                gross.append(None)
        else:
            names.append(f.find(class_="lister-item-header").get_text().replace('\n','').replace(' ',''))
            ratings.append(f.find(class_="inline-block ratings-imdb-rating").get_text().strip())
            metas.append(None)
            fvg=f.find(class_="sort-num_votes-visible").get_text().replace('\n','').replace(' ','').split("|")
            votes.append(fvg[0].split(':')[1].replace(',', ''))
            gross.append(None)            
                    
```

```python
len(names), len(ratings), len(metas), len(votes), len(gross)  
```

```python
d={"Movie Name":names, 'Ratings':ratings,'Metascores':metas,'Votes':votes,"Gross":gross}
df = pd.DataFrame(d)
df
```

```python
df.to_csv("IMDB_Movies.csv",index=False)
```
